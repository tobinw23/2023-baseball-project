```python
from pybaseball import statcast
statcast(start_dt="2024-03-28", end_dt="2024-04-13")
```

```python
def get_statcast_data_for_pitchers(pitcher_names):
  statcast_data = statcast(start_dt='2024-03-28', end_dt='2024-04-20')
  statcast_data_for_pitchers = statcast_data[statcast_data['player_name'].isin(pitcher_names)]
  return statcast_data_for_pitchers
```

```python
yankees_pitchers = ['Stroman, Marcus','Weaver, Luke', 'Cortes Jr., Nestor','Holmes, Clay','Ferguson, Caleb','Burdi, Nick','Rodon, Carlos','Schmidt, Clarke','Gil, Luis','Gonzalez, Victor','Santana, Dennis','Hamilton, Ian','Cosuins, Jake','Loaisiga, Jonathan', 'Poteet, Cody', 'Marinaccio, Ron','Beeter, Clayton','Maciejewski, Josh']
```

```python
statcast_data_for_yankees_pitchers = get_statcast_data_for_pitchers(yankees_pitchers)
```

```python
statcast_data_for_yankees_pitchers
```

```python
print(statcast_data_for_yankees_pitchers.columns.tolist())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(20, 10))

sns.barplot(x='pitch_type', y='release_speed', data=statcast_data_for_yankees_pitchers, hue='description', palette='Blues', edgecolor='grey', alpha=0.7, ax=ax)
for c in ax.containers:
    ax.bar_label(c)
plt.show()
```

```python
sns.relplot(x='release_speed', y='effective_speed', data=statcast_data_for_yankees_pitchers, hue=statcast_data_for_yankees_pitchers[['player_name','pitch_type']].apply(tuple,axis=1))
plt.show()
```

```python
import pandas as pd
import numpy as np
Nyy=statcast_data_for_yankees_pitchers.groupby(['player_name','pitch_type'])[['release_speed', 'release_pos_x', 'release_pos_z', 'pfx_x', 'pfx_z', 'plate_x', 'plate_z', 'hc_x', 'hc_y', 'hit_distance_sc', 'launch_speed', 'launch_angle', 'effective_speed', 'release_spin_rate', 'release_extension']].mean()
display(Nyy)
```

```python
Nyy.reset_index(inplace=True)
```

```python
Nyy.to_csv("Nyy.csv", index=False)
```

```python
Nyy.plot.scatter(x='release_speed', y='launch_speed')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.scatterplot(x='release_speed', y='launch_speed', data=Nyy, hue='pitch_type', ax=ax)

ax.set_xlabel('Release Speed (mph)', fontsize=14)
ax.set_ylabel('Launch Speed (mph)', fontsize=14)
ax.set_title('Scatterplot of Release Speed vs. Launch Speed for Yankees Pitchers', fontsize=16)

ax.legend(title='Pitch Type', fontsize=12)

plt.show()
```

```python
Nyy['launch_speed'].fillna(Nyy['launch_speed'].mean(), inplace=True)
fig, ax = plt.subplots(figsize=(10, 6))
sns.scatterplot(x='release_speed', y='launch_speed', data=Nyy, hue='pitch_type', ax=ax)

ax.set_xlabel('Release Speed (mph)', fontsize=14)
ax.set_ylabel('Launch Speed (mph)', fontsize=14)
ax.set_title('Scatterplot of Release Speed vs. Launch Speed for Yankees Pitchers', fontsize=16)

ax.legend(title='Pitch Type', fontsize=12, bbox_to_anchor=(1.05, 1), loc='lower left')

for i, row in Nyy.iterrows():
    ax.annotate(row['player_name'], (row['release_speed'], row['launch_speed']))

plt.show()
```
